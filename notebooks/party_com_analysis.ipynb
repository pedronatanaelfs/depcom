```python
import pandas as pd
```

```python
# Definir o caminho para o arquivo CSV
file_path = '../data/party_community_table.csv'

# Ler o arquivo CSV para um DataFrame
df = pd.read_csv(file_path)

df
```

```python
# Eliminar a linha 1 (que contém valores NaN)
df = df.drop(index=1).reset_index(drop=True)

# Definir a coluna "sigla_partido" como a primeira coluna (index)
df = df.drop(index=2)     # Remover a linha usada como cabeçalho

df = df.rename(columns={'year': 'party'})

df
```

```python
sufix = df.iloc[0, 1:].astype(int).astype(str)

# Remover o sufixo ".1", ".2", etc, das colunas (exceto a primeira)
cleaned_columns = [df.columns[0]] + [col.split('.')[0] for col in df.columns[1:]]

# Atualizar os nomes das colunas
df.columns = cleaned_columns

# Renomear as colunas apenas a partir da segunda coluna
new_columns = ['party'] + [f"{col}_{suf}" for col, suf in zip(df.columns[1:], sufix)]

# Aplicar os novos nomes ao DataFrame
df.columns = new_columns

# Remover a primeira linha (que continha os valores usados no sufixo)
df = df.drop(0)

df
```

```python
df.columns.tolist()
```

```python
# Extrair os nomes das colunas (a partir da segunda coluna) e remover o sufixo "_X"
years = [int(col.split('_')[0]) for col in df.columns[1:]]

# Encontrar o último ano (maior valor)
ultimo_ano = max(years)

print(f"O último ano no DataFrame é: {ultimo_ano}")
```

```python
def maior_partido_por_comunidade(df, ano):
    colunas_ano = [col for col in df.columns if col.startswith(str(ano))]
    print(f"\nAnalisando o ano {ano}, colunas do ano: {colunas_ano}")
    
    resultados = {}
    for col in colunas_ano:
        comunidade = col.split('_')[1]
        print(f"\nProcessando comunidade {comunidade} na coluna {col}")
        # Obter o partido com o maior número de deputados nesta comunidade
        total_por_partido = df.groupby('party')[col].sum()
        partido = total_por_partido.idxmax()
        deputados_partido_comunidade = total_por_partido.loc[partido]
        total_deputados_partido = df.loc[df['party'] == partido, colunas_ano].sum().sum()
        percentual_partido = (deputados_partido_comunidade / total_deputados_partido) * 100
        print(f"Partido com maior número de deputados na comunidade {comunidade}: {partido}")
        print(f"Deputados do partido na comunidade: {deputados_partido_comunidade}")
        print(f"Total de deputados do partido no ano: {total_deputados_partido}")
        print(f"Percentual do partido na comunidade: {percentual_partido}%")
        
        resultados[f'Comunidade_{comunidade}'] = {
            'partido': partido,
            'deputados': deputados_partido_comunidade,
            'percentual': round(percentual_partido, 2)
        }
    
    return resultados

def ajustar_rotulos_comunidades(df, ano_especificado):
    # Obter as maiores comunidades do ano especificado
    comunidades_ano_especificado = maior_partido_por_comunidade(df, ano_especificado)
    print(f"\nComunidades no ano especificado {ano_especificado}: {comunidades_ano_especificado}")
    
    # Obter os anos anteriores ao ano especificado
    anos_anteriores = sorted(
        set(int(col.split('_')[0]) for col in df.columns if col != 'party' and int(col.split('_')[0]) < ano_especificado),
        reverse=True
    )
    print(f"Anos anteriores ao ano especificado: {anos_anteriores}")
    
    for ano in anos_anteriores:
        colunas_ano_atual = [col for col in df.columns if col.startswith(str(ano))]
        print(f"\nAnalisando o ano {ano}, colunas do ano: {colunas_ano_atual}")
        
        # Ordenar as comunidades por número de deputados (do menor para o maior)
        comunidades_ordenadas = sorted(
            comunidades_ano_especificado.items(), key=lambda x: x[1]['deputados']
        )
        print(f"Comunidades ordenadas por número de deputados: {comunidades_ordenadas}")
        
        for comunidade_ref, info_ref in comunidades_ano_especificado.items():
            partido_ref = info_ref['partido']
            comunidade_num_ref = comunidade_ref.split('_')[1]
            print(f"\nReferência da comunidade: {comunidade_ref}")
            print(f"Partido de referência: {partido_ref}")
            print(f"Número da comunidade de referência: {comunidade_num_ref}")
            
            # Verificar em qual comunidade o partido_ref está no ano atual
            comunidade_encontrada = None
            for col in colunas_ano_atual:
                # Obter o total de deputados do partido na comunidade atual
                valor_partido_series = df.loc[df['party'] == partido_ref, col]
                print(f"Analisando partido {partido_ref} na coluna {col}: valor_partido_series = {valor_partido_series}")
                if not valor_partido_series.empty:
                    valor_partido = valor_partido_series.sum()
                    print(f"valor_partido somado: {valor_partido}")
                    if valor_partido > 0:
                        comunidade_encontrada = col.split('_')[1]
                        print(f"Partido {partido_ref} encontrado na comunidade {comunidade_encontrada} do ano {ano}")
                        break
                else:
                    print(f"Nenhum valor encontrado para o partido {partido_ref} na coluna {col}")
            
            # Se o partido não foi encontrado em nenhuma comunidade no ano atual
            if comunidade_encontrada is None:
                print(f"Partido {partido_ref} não encontrado em nenhuma comunidade no ano {ano}")
                # Vamos definir old_col como o nome da comunidade de referência
                old_col = f'{ano}_{comunidade_num_ref}'
                # Se old_col não existe, criamos a coluna com valores zero
                if old_col not in df.columns:
                    print(f"Coluna {old_col} não existe. Criando coluna com valores zero.")
                    df[old_col] = 0.0
                # Nenhuma troca é necessária, pois estamos apenas criando a coluna faltante
                continue
            
            old_col = f'{ano}_{comunidade_encontrada}'
            new_col = f'{ano}_{comunidade_num_ref}'
            
            # Criar new_col se não existir
            if new_col not in df.columns:
                print(f"Coluna {new_col} não existe. Criando coluna com valores zero.")
                df[new_col] = 0.0
            
            # Se old_col e new_col são iguais, não precisamos fazer nada
            if old_col == new_col:
                print(f"As colunas {old_col} e {new_col} já são iguais. Nenhuma ação necessária.")
                continue
            
            # Trocar os nomes das colunas
            print(f"Preparando para trocar nomes das colunas {old_col} e {new_col}")
            df.rename(columns={old_col: 'temp_col_name'}, inplace=True)
            df.rename(columns={new_col: old_col}, inplace=True)
            df.rename(columns={'temp_col_name': new_col}, inplace=True)
            print(f"Colunas após troca: {df.columns.tolist()}")
    
    return df

df_ajustado = ajustar_rotulos_comunidades(df, ultimo_ano)
df_ajustado
```

```python
def ordenar_colunas(df):
    # Excluir 'party' das colunas a serem ordenadas
    columns = [col for col in df.columns if col != 'party']
    # Extrair ano e comunidade de cada coluna
    col_info = []
    for col in columns:
        parts = col.split('_')
        if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
            year = int(parts[0])
            community = int(parts[1])
            col_info.append((col, year, community))
        else:
            # Colunas que não seguem o padrão 'ano_comunidade' ficam no final
            col_info.append((col, float('inf'), float('inf')))
    # Ordenar as colunas pelo ano e comunidade
    col_info_sorted = sorted(col_info, key=lambda x: (x[1], x[2]))
    # Obter os nomes das colunas ordenadas
    sorted_columns = ['party'] + [col[0] for col in col_info_sorted]
    # Reindexar o DataFrame com as colunas ordenadas
    df = df.reindex(columns=sorted_columns)
    return df
```

```python
# Ordenar as colunas do menor ano para o maior ano
df_ordenado = ordenar_colunas(df_ajustado)

df_ordenado
```
