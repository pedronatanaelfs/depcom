```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/detailed_results.csv')

df
```

```python
    # Initialize lists to hold data for plotting
years = []
mod_no_filter_no_prune = []  # For "Polarization 0-100, Prune 0%"
mod_filter_no_prune = []  # For "Max Modularity (Prune 0%)"
mod_no_filter_prune = []  # For "Max Modularity (Polarization 0-100)"
mod_both_methods = []  # For "Max Modularity Overall"

# Loop through each year to gather the necessary values
for year in df['Year'].unique():
    # Filter data for the current year
    df_year = df[df['Year'] == year]
    
    # 1. Modularity for Polarization 0-100 with Pruning 0%
    mod_no_filter_no_prune.append(
        df_year[(df_year['Polarization Lower Bound (%)'] == 0) & 
                (df_year['Polarization Upper Bound (%)'] == 100) & 
                (df_year['Pruning Percentage (%)'] == 0)]['Modularity'].values[0]
    )
    
    # 2. Maximum modularity for any interval of polarization, but with 0% pruning
    mod_filter_no_prune.append(
        df_year[df_year['Pruning Percentage (%)'] == 0]['Modularity'].max()
    )
    
    # 3. Maximum modularity for Polarization 0-100 regardless of pruning
    mod_no_filter_prune.append(
        df_year[(df_year['Polarization Lower Bound (%)'] == 0) & 
                (df_year['Polarization Upper Bound (%)'] == 100)]['Modularity'].max()
    )
    
    # 4. Maximum modularity overall (ignoring pruning and polarization interval)
    mod_both_methods.append(
        df_year['Modularity'].max()
    )
    
    # Append the year
    years.append(year)

# Plotting
x = range(len(years))  # x-axis positions

plt.figure(figsize=(14, 7))

# Plot four sets of bars with the requested names and order
plt.bar(x, mod_no_filter_no_prune, width=0.2, label='Modularity without Polarization Filter and Prone Optimization', align='center')
plt.bar([i + 0.2 for i in x], mod_filter_no_prune, width=0.2, label='Modularity with Polarization Filter applied but without Prone Optimization', align='center')
plt.bar([i + 0.4 for i in x], mod_no_filter_prune, width=0.2, label='Modularity with applied Prone Optimization but without Polarization Filter applied', align='center')
plt.bar([i + 0.6 for i in x], mod_both_methods, width=0.2, label='Modularity applying both methods', align='center')

# Add labels and title
plt.xticks([i + 0.3 for i in x], years, rotation=90)
plt.ylabel('Modularity')
plt.title('Modularity Comparison by Year with Different Filters and Optimization')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Initialize lists to hold data for plotting
years = []
mod_no_filter_no_prune = []  # For "Polarization 0-100, Prune 0%"
mod_filter_no_prune = []  # For "Max Modularity (Prune 0%)"
mod_no_filter_prune = []  # For "Max Modularity (Polarization 0-100)"
mod_both_methods = []  # For "Max Modularity Overall"

# Loop through each year to gather the necessary values
for year in df['Year'].unique():
    # Filter data for the current year
    df_year = df[df['Year'] == year]
    
    # 1. Modularity for Polarization 0-100 with Pruning 0%
    mod_no_filter_no_prune.append(
        df_year[(df_year['Polarization Lower Bound (%)'] == 0) & 
                (df_year['Polarization Upper Bound (%)'] == 100) & 
                (df_year['Pruning Percentage (%)'] == 0)]['Modularity'].values[0]
    )
    
    # 2. Maximum modularity for any interval of polarization, but with 0% pruning
    mod_filter_no_prune.append(
        df_year[df_year['Pruning Percentage (%)'] == 0]['Modularity'].max()
    )
    
    # Append the year
    years.append(year)

# Plotting
x = range(len(years))  # x-axis positions

plt.figure(figsize=(14, 7))

# Plot four sets of bars with the requested names and order
plt.bar(x, mod_no_filter_no_prune, width=0.2, label='Modularity without Polarization Filter and Prone Optimization', align='center')
plt.bar([i + 0.2 for i in x], mod_filter_no_prune, width=0.2, label='Modularity with Polarization Filter applied but without Prone Optimization', align='center')

# Add labels and title
plt.xticks([i + 0.3 for i in x], years, rotation=90)
plt.ylabel('Modularity')
plt.title('Modularity Comparison by Year with Different Filters and Optimization')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Converting the lists to NumPy arrays for easier calculations
mod_no_filter_no_prune = np.array(mod_no_filter_no_prune)
mod_filter_no_prune = np.array(mod_filter_no_prune)
mod_no_filter_prune = np.array(mod_no_filter_prune)
mod_both_methods = np.array(mod_both_methods)

# 1. Improvement after applying polarization filter, without pruning optimization
improvement_filter = ((mod_filter_no_prune - mod_no_filter_no_prune) / mod_no_filter_no_prune) * 100
avg_improvement_filter = np.mean(improvement_filter)
max_improvement_filter = np.max(improvement_filter)
year_max_filter = years[np.argmax(improvement_filter)]

# 2. Improvement after applying pruning optimization, without polarization filter
improvement_prune = ((mod_no_filter_prune - mod_no_filter_no_prune) / mod_no_filter_no_prune) * 100
avg_improvement_prune = np.mean(improvement_prune)
max_improvement_prune = np.max(improvement_prune)
year_max_prune = years[np.argmax(improvement_prune)]

# 3. Improvement with both methodologies applied
improvement_both = ((mod_both_methods - mod_no_filter_no_prune) / mod_no_filter_no_prune) * 100
avg_improvement_both = np.mean(improvement_both)
max_improvement_both = np.max(improvement_both)
year_max_both = years[np.argmax(improvement_both)]

# 4. Improvement with both methodologies applied compared to only pruning optimization
improvement_both_vs_prune = ((mod_both_methods - mod_no_filter_prune) / mod_no_filter_prune) * 100
avg_improvement_both_vs_prune = np.mean(improvement_both_vs_prune)
max_improvement_both_vs_prune = np.max(improvement_both_vs_prune)
year_max_both_vs_prune = years[np.argmax(improvement_both_vs_prune)]

# Printing results
print(f"1. Average improvement with polarization filter applied (no pruning optimization): {avg_improvement_filter:.2f}%")
print(f"   Maximum improvement in {year_max_filter} with {max_improvement_filter:.2f}% improvement")

print(f"\n2. Average improvement with pruning optimization (no polarization filter): {avg_improvement_prune:.2f}%")
print(f"   Maximum improvement in {year_max_prune} with {max_improvement_prune:.2f}% improvement")

print(f"\n3. Average improvement with both methodologies applied: {avg_improvement_both:.2f}%")
print(f"   Maximum improvement in {year_max_both} with {max_improvement_both:.2f}% improvement")

print(f"\n4. Average improvement with both methodologies compared to only pruning optimization: {avg_improvement_both_vs_prune:.2f}%")
print(f"   Maximum improvement in {year_max_both_vs_prune} with {max_improvement_both_vs_prune:.2f}% improvement")
```
